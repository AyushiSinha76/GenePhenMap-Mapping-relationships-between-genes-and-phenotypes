```python
import random

# Function to assign causal and non-causal labels
def assign_causal_noncausal(gene_string):
    # Convert the gene string into a list of genes
    genes = gene_string.strip('{}').split(',')
    
    # Randomly assign one gene as causal and the rest as non-causal
    random.shuffle(genes)
    causal_gene = genes[0]
    non_causal_genes = genes[1:]
    
    return causal_gene, non_causal_genes

# Apply the function to assign causal and non-causal labels
sampled_phenotypes['causal_gene'], sampled_phenotypes['non_causal_genes'] = zip(
    *sampled_phenotypes['symbol_gene_string'].apply(assign_causal_noncausal)
)

# Display the phenotype with causal and non-causal gene mappings
print(sampled_phenotypes[['description', 'causal_gene', 'non_causal_genes']].head())
```

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import ast

# Load gene and phenotype embeddings (if not already loaded)
gene_embeddings = pd.read_csv('C:/Users/ashut/Downloads/11391053/zenodo_directory/zenodo_directory/data/helper_datasets/gene_embeddings.csv')
phenotype_embeddings = pd.read_csv('C:/Users/ashut/Downloads/11391053/zenodo_directory/zenodo_directory/data/helper_datasets/phenotype_embeddings.csv')

# Convert the string representation of lists in the 'embedding' column to actual lists
gene_embeddings['embedding'] = gene_embeddings['embedding'].apply(ast.literal_eval)
phenotype_embeddings['embedding'] = phenotype_embeddings['embedding'].apply(ast.literal_eval)

# Convert the list of embeddings to a NumPy array (2D array)
gene_embeddings_numeric = np.vstack(gene_embeddings['embedding'].values)
phenotype_embeddings_numeric = np.vstack(phenotype_embeddings['embedding'].values)

# Apply PCA
pca = PCA(n_components=2)

# Fit PCA on gene and phenotype embeddings
gene_pca = pca.fit_transform(gene_embeddings_numeric)
phenotype_pca = pca.fit_transform(phenotype_embeddings_numeric)

# Plot PCA results
plt.scatter(gene_pca[:, 0], gene_pca[:, 1], label='Genes', alpha=0.5)
plt.scatter(phenotype_pca[:, 0], phenotype_pca[:, 1], label='Phenotypes', alpha=0.5)
plt.legend()
plt.title("PCA of Gene and Phenotype Embeddings")
plt.show()
```

```python
from scipy.spatial import distance_matrix

# Calculate the distance matrix between gene embeddings and phenotype embeddings
distances = distance_matrix(gene_pca, phenotype_pca)

# Create a DataFrame for easy analysis
distance_df = pd.DataFrame(distances, index=gene_embeddings.index, columns=phenotype_embeddings.index)

# Find the phenotype with the smallest distance for each gene
closest_phenotypes = distance_df.idxmin(axis=1)

# Create a DataFrame showing each gene and its closest phenotype
closest_matches = pd.DataFrame({
    'Gene_Index': gene_embeddings.index,
    'Closest_Phenotype_Index': closest_phenotypes,
    'Distance': distance_df.min(axis=1)
})

# Display the closest matches
print(closest_matches.head())
```

```python
plt.scatter(gene_pca[:, 0], gene_pca[:, 1], label='Genes', alpha=0.5)
plt.scatter(phenotype_pca[:, 0], phenotype_pca[:, 1], label='Phenotypes', alpha=0.5)

# Draw lines between the closest pairs
for row in closest_matches.itertuples(index=False):
    gene_idx = row.Gene_Index
    phenotype_idx = row.Closest_Phenotype_Index
    
    plt.plot([gene_pca[gene_idx, 0], phenotype_pca[phenotype_idx, 0]],
             [gene_pca[gene_idx, 1], phenotype_pca[phenotype_idx, 1]], color='gray', alpha=0.3)

plt.legend()
plt.title("PCA of Gene and Phenotype Embeddings with Closest Matches")
plt.show()
```

```python
from sklearn.cluster import KMeans

# Set the number of clusters (you can try different values)
num_clusters = 5

# Apply K-Means clustering on the PCA-reduced gene embeddings
kmeans_genes = KMeans(n_clusters=num_clusters, n_init=10, random_state=0).fit(gene_pca)
gene_labels = kmeans_genes.labels_

# Apply K-Means clustering on the PCA-reduced phenotype embeddings
kmeans_phenotypes = KMeans(n_clusters=num_clusters, n_init=10, random_state=0).fit(phenotype_pca)
phenotype_labels = kmeans_phenotypes.labels_

# Plot the clusters for genes and phenotypes
plt.scatter(gene_pca[:, 0], gene_pca[:, 1], c=gene_labels, cmap='viridis', label='Genes', alpha=0.5)
plt.scatter(phenotype_pca[:, 0], phenotype_pca[:, 1], c=phenotype_labels, cmap='plasma', label='Phenotypes', alpha=0.5)
plt.legend()
plt.title("K-Means Clustering of Gene and Phenotype Embeddings")
plt.show()
```

```python
from sklearn.metrics import silhouette_score
gene_silhouette = silhouette_score(gene_pca, gene_labels)
phenotype_silhouette = silhouette_score(phenotype_pca, phenotype_labels)
print(f'Gene Silhouette Score: {gene_silhouette}')
print(f'Phenotype Silhouette Score: {phenotype_silhouette}')
```

```python
inertia_values_genes = []
inertia_values_phenotypes = []

for num_clusters in range(2, 11):
    kmeans_genes = KMeans(n_clusters=num_clusters, n_init=10, random_state=0).fit(gene_pca)
    kmeans_phenotypes = KMeans(n_clusters=num_clusters, n_init=10, random_state=0).fit(phenotype_pca)
    
    inertia_values_genes.append(kmeans_genes.inertia_)
    inertia_values_phenotypes.append(kmeans_phenotypes.inertia_)

# Plot the Elbow Graph for Genes and Phenotypes
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(range(2, 11), inertia_values_genes, marker='o')
plt.title('Elbow Method for Genes')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')

plt.subplot(1, 2, 2)
plt.plot(range(2, 11), inertia_values_phenotypes, marker='o')
plt.title('Elbow Method for Phenotypes')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')

plt.tight_layout()
plt.show()
```
